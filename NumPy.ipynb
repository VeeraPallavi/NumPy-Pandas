```python
np.random.seed(2)
marks = np.random.randint(0, 100, size = (200, 6))
```

```python
total_scores = np.sum(marks, axis = 1)
total_scores
```

```python
percentage = (total_scores / 600) * 100
np.round(percentage, 3)
```

```python
#Assign grades using vectorized operations
grades = np.where(percentage >= 90, 'A', 
         np.where(percentage >= 75, 'B',
         np.where(percentage >= 60, 'C',
         np.where(percentage >= 40, 'D', 'F'))))
print(grades)
```

```python
#Identify top 10% performers
cut_off = np.percentile(percentage, 90)
top_10_performers = percentage >= cut_off
percentage[top_10_performers]
```

```python
np.random.seed(3)
temparature = np.random.randint(-20, 50, size=(7,24))
temparature
```

```python
#Find the daily maximum and minimum temperature
maximum_temparatures = temparature.max(axis = 1)
minimum_temparatures = temparature.min(axis = 1)
print(f"Maximum Temperatures : {maximum_temparatures}")
print(f"Minimum Temperatures : {minimum_temparatures}")
```

```python
daily_variation = temparature.max(axis=1) - temparature.min(axis=1)
print(f"Daily Temparature variations : {daily_variation}")
```

```python
#Identify the day with the largest temperature variation
largest_temparature_variation = np.argmax(daily_variation)
print(f"Day with largest temparature variation : {largest_temparature_variation}")
```

```python
#Replace outliers beyond ±2 standard deviations with the mean temperature
mean_temperature = temparature.mean()
std_deviation = temparature.std()
print(mean_temperature)
print(std_deviation)
```

```python
lower_bound = mean_temperature - 2 * std_deviation
upper_bound = mean_temperature + 2 * std_deviation
print(lower_bound)
print(upper_bound)
```

```python
cleaned_temperature = np.where((temparature < lower_bound) | (temparature > upper_bound), mean_temperature, temparature)
print(cleaned_temperature)
```

```python
np.random.seed(0)
transaction = np.random.randint(-5000, 5000, size = (20, 50))
transaction
```

```python
#Separate debit and credit transactions
credits = np.where(transaction > 0, transaction, 0)
debits = np.where(transaction < 0, transaction, 0)
print("Credits : ")
print(credits)
print("Debits : ")
print(debits)
```

```python
#Compute daily net balance changes
net_balance = transaction.sum(axis = 1)
net_balance
```

```python
#Identify accounts with continuous negative balance trends
cumulative_balance = np.cumsum(transaction, axis=0)
daily_diff = np.diff(cumulative_balance, axis=0)
continuous_negative = np.all(daily_diff < 0, axis=0)
negative_trend_accounts = np.where(continuous_negative)[0]
print("Accounts with continuous negative balance trend:")
print(negative_trend_accounts)
```

```python
np.random.seed(0)
heart_rate = np.random.randint(50, 120, size=(60, 24)).astype(float)

# Introduce missing values
heart_rate[heart_rate < 55] = np.nan

heart_rate
```

```python
#Compute hourly average heart rate
hourly_avg = np.nanmean(heart_rate, axis=0)
print("Hourly average heart rate:")
print(hourly_avg)
```

```python
#Detect abnormal readings outside the normal range
abnormal_readings = (heart_rate < 60) | (heart_rate > 100)
print("Abnormal readings (True = abnormal):")
print(abnormal_readings)
```

```python
#Replace missing values using forward fill logic
filled_heart_rate = heart_rate.copy()

for i in range(1, filled_heart_rate.shape[1]):
    mask = np.isnan(filled_heart_rate[:, i])
    filled_heart_rate[mask, i] = filled_heart_rate[mask, i - 1]

print("Heart rate after forward fill:")
print(filled_heart_rate)
```

```python
np.random.seed(0)
requests = np.random.poisson(lam=200, size=1440).reshape(24, 60)
requests
```

```python
#Identify peak traffic hours
hourly_total = requests.sum(axis=1)
peak_hour = np.argmax(hourly_total)
print("Peak traffic hour:", peak_hour)
```

```python
#Detect sudden traffic spikes
mean_req = requests.mean()
std_req = requests.std()
spike_threshold = mean_req + 3 * std_req
spikes = np.where(requests > spike_threshold)[0]
print("Spike indices (minutes):")
print(spikes[:10])
```
